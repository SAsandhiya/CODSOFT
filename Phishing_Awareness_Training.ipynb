```python
import time

# --- Presentation Content Data ---
PHISHING_MODULE = {
    "title": "ðŸŽ£ Phishing Awareness Training: Don't Take the Bait!",
    "sections": [
        {
            "heading": "What is Phishing?",
            "content": [
                "Phishing is a social engineering attack where an attacker impersonates a trustworthy entity to trick victims into providing sensitive data, typically login credentials or financial information.",
                "It's usually carried out via email, but can also happen via SMS ('smishing') or voice calls ('vishing')."
            ]
        },
        {
            "heading": "ðŸ” How to Recognize a Phishing Email",
            "content": [
                "**1. Sender's Email Address:** Check the actual email address, not just the display name. Look for subtle misspellings (e.g., 'Aazon.com' instead of 'Amazon.com').",
                "**2. Urgent or Threatening Language:** Attackers often use fear, urgency, or extreme curiosity to rush you into acting without thinking (e.g., 'Your account will be suspended in 2 hours!').",
                "**3. Suspicious Links/Attachments:** **NEVER** click a link without hovering over it first to check the destination URL. Does the link match the text? Are you expecting the attachment?",
                "**4. Poor Grammar and Spelling:** Legitimate companies have professional communications teams. Mistakes are a major red flag.",
                "**5. Generic Greetings:** 'Dear Customer' is a common sign of a mass phishing attack. Spear phishing (highly targeted attacks) is often personalized, but generic greetings are a common clue."
            ]
        },
        {
            "heading": "ðŸ›¡ï¸ Best Practices to Avoid Falling Victim",
            "content": [
                "**Hover Before You Click:** Always hover your mouse over links to preview the URL in the bottom-left of your browser/email client.",
                "**Verify, Verify, Verify:** If an email from 'HR' asks you to click a link to update your payroll, ignore the email and contact HR directly through a known, trusted channel (like an internal phone number).",
                "**Use Multi-Factor Authentication (MFA):** Even if an attacker steals your password, MFA prevents them from logging in.",
                "**Report Suspicious Activity:** Notify your IT or Security team immediately about any suspicious emails."
            ]
        }
    ]
}

# --- Quiz Data ---
QUIZ_QUESTIONS = [
    {
        "question": "Which of these is the **BEST** way to check if a link in an email is safe?",
        "options": {
            "A": "Click the link and see if the website looks official.",
            "B": "Copy the link and paste it into a search engine.",
            "C": "Hover your mouse cursor over the link to preview the destination URL."
        },
        "answer": "C",
        "explanation": "Hovering over the link allows you to see the true destination URL in the status bar without clicking, which is the safest first step."
    },
    {
        "question": "An email from 'Microsoft Security' tells you your account will be **locked immediately** unless you log in now. This is an example of what social engineering tactic?",
        "options": {
            "A": "Baiting",
            "B": "Creating a sense of Urgency/Fear",
            "C": "Pretexting"
        },
        "answer": "B",
        "explanation": "Creating urgency or fear rushes the victim into making a mistake and prevents them from thinking critically about the email's legitimacy."
    },
    {
        "question": "You receive an email from 'PayPaI' (with a capital 'i' instead of an 'l') asking you to verify your address. What is the most obvious red flag?",
        "options": {
            "A": "It asks you to verify your address.",
            "B": "The logo is slightly old.",
            "C": "A subtle misspelling in the sender's name/domain."
        },
        "answer": "C",
        "explanation": "Typosquatting or subtle misspellings in the sender's domain or name is a classic phishing trick used to fool a quick glance."
    }
]

# --- Functions ---

def display_section(heading, content):
    """Displays a single section of the training module."""
    print(f"\n{'='*50}")
    print(f"### {heading} ###".center(50))
    print('='*50)
    for point in content:
        print(f"* {point}")
    time.sleep(1) # Pause for readability

def run_quiz(questions):
    """Runs the interactive multiple-choice quiz."""
    print("\n" + "="*50)
    print("### ðŸ“ Interactive Phishing Quiz ###".center(50))
    print('='*50)
    score = 0
    total = len(questions)

    for i, q in enumerate(questions):
        print(f"\n--- Question {i + 1} of {total} ---")
        print(q["question"])
        for option, text in q["options"].items():
            print(f"  [{option}] {text}")

        # Input loop to ensure valid choice
        while True:
            user_answer = input("Your answer (A, B, or C): ").strip().upper()
            if user_answer in q["options"]:
                break
            print("Invalid input. Please choose A, B, or C.")

        if user_answer == q["answer"]:
            print("âœ… Correct! Well done.")
            score += 1
        else:
            print(f"âŒ Incorrect. The correct answer was [{q['answer']}].")
            print(f"   Explanation: {q['explanation']}")
        time.sleep(1.5)

    print("\n" + "="*50)
    print("### Quiz Results ###".center(50))
    print('='*50)
    print(f"You answered {score} out of {total} questions correctly.")
    if score == total:
        print("Excellent work! You are a Phishing Awareness Expert! â­")
    elif score >= total / 2:
        print("Good effort! You know the basics, but review the section on suspicious links.")
    else:
        print("Keep practicing! Review all sections to strengthen your knowledge.")
    print('='*50)

def main():
    """Main function to run the phishing awareness training."""
    print("\n" + "#" * 60)
    print(PHISHING_MODULE["title"].center(60))
    print("#" * 60)
    time.sleep(1)

    # Display the training sections
    for section in PHISHING_MODULE["sections"]:
        display_section(section["heading"], section["content"])

    # Run the interactive quiz
    run_quiz(QUIZ_QUESTIONS)

    print("\n--- Training Complete ---")

if __name__ == "__main__":
    main()
```
